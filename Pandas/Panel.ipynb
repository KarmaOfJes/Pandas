## Transpose a `Panel` with the `.transpose()` Method

```python
companies = ["MSFT", "GOOG", "AAPL", "YHOO", "AMZN"]

p = data.DataReader(name = companies, data_source = "yahoo", start = "2010-01-01", end = "2016-12-31")

p
```

```python
p.axes
```

```python
p2 = p.T
p2
```

```python
p2.loc[('Close', 'GOOG')][:]
```

```python
p2.loc[[('Close', 'GOOG'), ('Open', 'GOOG')]][:]
```

```python
p2.loc[(slice(None), 'GOOG'), :]
```

```python
p.loc[:, (slice(None), 'GOOG')]
```

```python
p2.axes[0].names
```

```python
p2.axes[0].get_level_values(0), p2.axes[0].get_level_values(1)
```

```python
p.columns.levels
```

## The `.swapaxes()` Method

```python
companies = ["MSFT", "GOOG", "AAPL", "YHOO", "AMZN"]

p = data.DataReader(name = companies, data_source = "yahoo", start = "2010-01-01", end = "2016-12-31")

p
```

```python
p.swaplevel(0,1, axis=1).head()
```

```python
p.T.swaplevel(0,1, axis = 0).head()
```

```python
p.reorder_levels([1,0], axis = 1).head()
```

```python
p.rename(columns = {'Adj Close': 'New name'}).head()
```

```python
p.T.rename(index = {'Adj Close': 'New name', 'GOOG': 'google', 'AAPL': 'apple'}).head()
```

```python
p.T.head(10).sort_index(level=[1,0])
```
