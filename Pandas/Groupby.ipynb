# Custom function to groupby()

```python
df = pd.DataFrame( { 'company': [ 'Alpha', 'Alpha', 'Alpha',
                                  'Beta', 'Beta', 'Beta',
                                  'Gamma', 'Gamma', 'Gamma'],
                     'region':  [ 'NE', 'NW', 'SW', 'NW', 'SW',
                                  'NE', 'NE', 'SW', 'NW'],
                     'revenue': [ 10, 9, 2, 15, 8, 2, 16, 3, 9],
                     'revtype': [ 'retail', 'retail', 'wholesale',
                                  'wholesale', 'wholesale', 'retail',
                                  'wholesale', 'retail', 'retail'     ] } )

def get_sum(df_slice):
    return sum(df_slice['revenue'])

print(df.groupby('region').apply(get_sum))  # custom function: same as groupby('region').sum()
```

```python
# Also you can pass function that will aggregate groups in goupby()
# NOTE: By default groupby() takes index of row as an argument

def bydecplace(idx):
    row = df.loc[idx]                 # a Series with the row values for this index
    return(len(str(row['revenue'])))  # '2' if 10; '1' if 9

print(df.groupby(bydecplace).sum())   # in groupby() it iteratively runs bydecplace() for eanh row, than rows are grouped
                                      # by particular values which were retrieved after call of bydecplace().
                                      # as groupby() took bydecplace() as argument, index was passed to bydecplace().
```

# Transform() and apply() approach to count minmax

```python
test_df = pd.DataFrame(
    {
        'A': [1, 1, 1, 2, 2, 2],
        'B': [1, 2, 3, 10, 20, 30],
        'C': [1, 2, 3, 10, 20, 30]
    }
)

test_df.assign(
    min_max = 
        lambda df: df.groupby('A')['B'].transform(lambda x: x.apply(lambda y: (x[x > y]).min()))
)
```

```python
# Here you see that For column C for value 200 next minmax should be 300, but it is 100, as in expression
# x.loc[x.B > y.B].min(), we pass [True, False] list which is applied ti column B, and for be the first row index is
# x.B = 2, which os x.C = 100
test_df.groupby('A').apply(lambda x: x.apply(lambda y: x.loc[x.B > y.B].min(), axis=1))
```

```python
# Here we don't reference to B as x.B or y.B, as we pass Series object after groupby('A')['B']
test_df.assign(
    min_max = 
        lambda df: df.groupby('A')['B'].apply(lambda x: x.apply(lambda y: x.loc[x > y].min()))
)
# Here we must take 'B' in [] if we want to select column 'B': x.loc[x['B'] > y['B'], ['B']], because taking column in[]
# returns Series output, instead of numpy.int scalrar, hence we will be able to assign new column as old indexies would
# be preserved
test_df.assign(
    min_max = 
        lambda df: df.groupby('A').apply(lambda x: x.apply(lambda y: x.loc[x['B'] > y['B'], ['B']].min(), axis = 1))
)
```
